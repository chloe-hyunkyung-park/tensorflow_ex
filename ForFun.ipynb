```python
data = np.array(win + lose)
print(data[0])
data = (data - data.min())/(data.max() - data.min())
#data = data.sample(frac=1)
print(data[0])
print(np.ceil(data[0] * 45))
```

```python
X = data
Y = np.array([1]*len(win) + [0]*len(lose))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=64, activation='relu', input_shape=(6,)),
    tf.keras.layers.Dense(units=32, activation='sigmoid'),
    tf.keras.layers.Dense(units=16, activation='relu'),
    tf.keras.layers.Dense(units=8, activation='sigmoid'),
    tf.keras.layers.Dense(units=4, activation='sigmoid'),
    tf.keras.layers.Dense(units=1, activation='relu')
])

model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.05), loss='mse')

model.summary()
```

```python
history = model.fit(X, Y, epochs=50, batch_size=32, validation_split=0.25)
#model.predict(np.array([[9, 10, 13, 28, 38, 45]]))
#model.predict(np.array([[2, 4, 11, 28, 29, 43]]))
```
